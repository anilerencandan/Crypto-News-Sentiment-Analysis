```python
import pandas as pd
import re

# CSV dosyasını yükle
df = pd.read_csv('/Users/anilcandan/Desktop/Yazlab-Scraping/csv_files/filtered_crypto_news.csv')

# Coin terimlerini tanımla
coin_terms = {'bitcoin', 'ethereum', 'solona', 'btc', 'eth', 'doge', 'sol', 'coin','openai', 'chatgpt4','chatgpt','chat','gpt'}

# Türkçe stop words listesi (örnek)
# Adım 1: Tüm verileri küçük harfe çevirme
df['İçerik'] = df['İçerik'].str.lower()
df['Başlık'] = df['Başlık'].str.lower()


def replace_punctuation(text):
    # 1. İki kelime arasında bulunan "-" işaretini tek boşlukla değiştir
    text = re.sub(r'(?<=\w)-(?=\w)', ' ', text)  # Kelimeler arasında "-" varsa, tek boşlukla değiştir
    
    # 2. Diğer tüm noktalama işaretlerini kaldır
    text = re.sub(r"[^\w\s]", '', text)
    
    return text

# Uygulama: İçerik ve Başlık sütunlarına işlemi uygula
df['İçerik'] = df['İçerik'].apply(lambda x: replace_punctuation(str(x)))
df['Başlık'] = df['Başlık'].apply(lambda x: replace_punctuation(str(x)))

df.head(50)
df.to_csv('/Users/anilcandan/Desktop/Yazlab-Scraping/csv_files/naber1.csv', index=False)



```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

# NLTK Türkçe stop words ve tokenizer'ı indirme (ilk kez çalıştırılıyorsa)
nltk.download('stopwords')
nltk.download('punkt')

# Türkçe stop words listesini ayarla
stop_words = set(stopwords.words('turkish'))
stop_words.add("bir")  # "bir" kelimesini özellikle çıkarmak için ekledik

# CSV dosyasını yükle
df = pd.read_csv('/Users/anilcandan/Desktop/Yazlab-Scraping/csv_files/processed1113.csv')

# Stop words temizleme fonksiyonu
def remove_stop_words(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# 'İçerik_kök1' ve 'Başlık_kök1' sütunlarından stop words kelimelerini çıkar
df['İçerik_kök_cleaned'] = df['İçerik_kök1'].apply(lambda x: remove_stop_words(str(x)))
df['Başlık_kök_cleaned'] = df['Başlık_kök1'].apply(lambda x: remove_stop_words(str(x)))

# Tüm filtrelenmiş kelimeleri bir liste olarak topla
all_filtered_words = ' '.join(df['İçerik_kök_cleaned']) + ' ' + ' '.join(df['Başlık_kök_cleaned'])
all_filtered_words = word_tokenize(all_filtered_words)

# Anahtar kelimelerin sayısını hesapla
word_count = Counter(all_filtered_words)

# En sık geçen 5 kelimeyi seç
labels = [word[0] for word in word_count.most_common(5)]
values = [word[1] for word in word_count.most_common(5)]

# Grafiği hazırla ve göster
plt.bar(labels, values)
plt.xlabel('Kelime')
plt.ylabel('Sıklık')
plt.title('En Sık Geçen 5 Kelime')
plt.show()


df.to_csv('/Users/anilcandan/Desktop/Yazlab-Scraping/csv_files/processed1115.csv', index=False)

```

```python
import pandas as pd

# CSV dosyasını yükle
df = pd.read_csv('/Users/anilcandan/Desktop/Yazlab-Scraping/csv_files/processed1115.csv')

# Coin terimlerini tanımla (bozulmadan kalması gereken kelimeler)
coin_terms = {'bitcoin', 'ethereum', 'solona', 'btc', 'eth', 'doge', 'sol', 'coin', 'openai', 'chatgpt'}

# Basit bir kök bulma fonksiyonu
def simple_turkish_stemmer(word):
    suffixes = ['lar', 'ler', 'ci', 'cı', 'de', 'da', 'ler', 'lar', 'in', 'un', 'im', 'um', 'sin', 'lık', 'lik', 'lar', 'ler', 'cı', 'ci', 'li', 'lı']
    for suffix in suffixes:
        if word.endswith(suffix) and word not in coin_terms:  # Coin terimlerini kök ayırma işlemine sokmadan korur
            return word[:-len(suffix)]
    return word

# Metni köklerine ayırma fonksiyonu
def stem_text(text):
    words = text.split()
    stemmed_words = []
    for word in words:
        if word in coin_terms:  # Coin terimlerini olduğu gibi koru
            stemmed_words.append(word)
        else:
            stemmed_words.append(simple_turkish_stemmer(word))  # Diğer kelimelerin kökünü bul
    return ' '.join(stemmed_words)

# 'İçerik_kök_cleaned' ve 'Başlık_kök_cleaned' sütunlarında kök bulma işlemi
df['İçerik_kök_rooted'] = df['İçerik_kök_cleaned'].apply(lambda x: stem_text(str(x)))
df['Başlık_kök_rooted'] = df['Başlık_kök_cleaned'].apply(lambda x: stem_text(str(x)))

# Sonucu yeni bir dosyaya kaydet
df.to_csv('/Users/anilcandan/Desktop/Yazlab-Scraping/csv_files/processed1116.csv', index=False)
print(df[['İçerik_kök_cleaned', 'İçerik_kök_rooted', 'Başlık_kök_cleaned', 'Başlık_kök_rooted']].head())
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
import nltk

# İlk kez çalıştırılıyorsa, gerekli nltk veri paketini indir
nltk.download('punkt')

# Türkçe kelime listesi dosyasını yükle
# Belirli bir dosya yolundan turkish_words.txt dosyasını okuma
file_path = '/Users/anilcandan/Desktop/Yazlab-Scraping/words.txt'

with open(file_path, 'r', encoding='utf-8') as f:
    turkish_words = set(f.read().splitlines())


# Kripto terimleri tanımla (korunması gereken terimler)
coin_terms = {'bitcoin', 'ethereum', 'solona', 'btc', 'eth', 'doge', 'sol', 'coin', 'openai', 'chatgpt'}


def simple_turkish_stemmer(word):
    # Kelimenin Türkçe olup olmadığını kontrol et
    if word in turkish_words or word in coin_terms:
        return word  # Eğer kelime zaten Türkçedeyse veya kripto terimlerindeyse olduğu gibi bırak
    suffixes = ['dan', 'den', 'lar', 'ler', 'ci', 'cı', 'de', 'da', 'in', 'un', 'im', 'um', 'sin', 'lık', 'lik', 'cı', 'ci', 'li', 'lı']
    for suffix in suffixes:
        if word.endswith(suffix):
            root = word[:-len(suffix)]
            if root in turkish_words:  # Eğer kök de Türkçeyse, kök olarak bırak
                return root
    return word  # Kök Türkçede değilse, olduğu gibi bırak

# Metni köklerine ayırma ve yalnızca Türkçe kökleri koruma fonksiyonu
def process_text(text):
    words = word_tokenize(text.lower())  # Metni küçük harfe çevir ve tokenlara ayır
    processed_words = []
    for word in words:
        stemmed_word = simple_turkish_stemmer(word)
        # Eğer kök Türkçe veya kripto terimiyse ekle
        if stemmed_word in turkish_words or stemmed_word in coin_terms:
            processed_words.append(stemmed_word)
    return ' '.join(processed_words)


# CSV dosyasını yükle
df = pd.read_csv('/Users/anilcandan/Desktop/Yazlab-Scraping/csv_files/processed1115.csv')

# 'İçerik_kök_cleaned' ve 'Başlık_kök_cleaned' sütunlarına Türkçe kontrolü uygula
df['İçerik_kök_rooted'] = df['İçerik_kök_cleaned'].apply(lambda x: process_text(str(x)))
df['Başlık_kök_rooted'] = df['Başlık_kök_cleaned'].apply(lambda x: process_text(str(x)))

# Sonucu yeni bir dosyaya kaydet
df.to_csv('processed1115221212121_with_turkish_words.csv', index=False)
print(df[['İçerik_kök_cleaned', 'İçerik_kök_rooted', 'Başlık_kök_cleaned', 'Başlık_kök_rooted']].head())
```
